Предположим, что m = 0.1 кг, k = 4 Н/м, b = 0.2 кг/с, x0 = 0.05 м и
v0 = 0 м/с. Постройте график зависимости x(t) в диапазоне времени от 0 до 5
секунд. Определите период колебаний T по графику. Сравните измеренный период
с рассчитанным теоретическим значением. Рассчитайте энергию, рассеянную в виде
тепла из-за сопротивления воздуха, за первые 2 секунды движения.

```python
m = 0.1
k = 4
b = 0.2
x0 = 0.05
v0 = 0
```

```python
beta = b / (2 * m)
w0 = np.sqrt(k / m)
time = np.arange(0, 5.1, 0.05)
time
```

```python
w = np.sqrt(w0 ** 2 - beta ** 2)
w
```

```python
x = np.exp(-beta * time) * (x0 * np.cos(w * time) + ((v0 + beta * x0) / w0) * np.sin(w * time))
x
```

```python
plt.title("x(t)")
plt.grid("light")
plt.plot(time, x)
plt.ylabel("x, м")
plt.xlabel("t, с")
plt.axvline(x=1, color='red', linestyle='-', label='Линии для T')
plt.axvline(x=2, color='red', linestyle='-', label='Линии для T')
plt.legend()
```

Практический период около секунды сходится с теоритическим.

Теперь считаем потерю энергии за 2 секунды

```python
beta = b / (2 * m)
w0 = np.sqrt(k / m)
time = np.arange(0.05, 2.05, 0.05)
w = np.sqrt(w0 ** 2 - beta ** 2)
x = np.exp(-beta * time) * (x0 * np.cos(w * time) + ((v0 + beta * x0) / w0) * np.sin(w * time))
```

```python
# здесь A - работа
v = np.abs(x / time)
A = b * v[v != np.inf] * 0.05
print(f"Рассеяная энергия: {np.round(sum(A) * 1e3, 1)} мДж")
```

```python
plt.title("A(t)")
plt.grid("light")
plt.ylabel("А, Дж")
plt.xlabel("t, с")
plt.plot(time, A)
```

Предположим, что m = 0.1 кг, k = 4 Н/м, b = 0.2 кг/с, x0 = 0 м и v0 = 5 м/с.
На расстоянии d = 0.02 м от положения равновесия шара расположена стенка, удары
о которую можно считать абсолютно упругими. Сколько ударов о стенку произойдет
в ходе движения шара?

```python
x0 = 0
v0 = 5
m = 0.1
k = 4
b = 0.2
d = 0.02
```

```python
beta = b / (2 * m)
w0 = np.sqrt(k / m)
dt = 1e-4
time = np.arange(0, 2 + dt, dt)
```

Столкновение о стенку происходит там, где x > d. Это будет продолжаться пока амплитуда больше d.

```python
# количество соударений со стенкой
cnt = 0
# фиксируем координаты и скорость
coord_x = []
velocity_x = []
# текущие данные
E = []
cur_x, cur_v = x0, v0
# перебираем значения по времени
for i in time:
    # заполнение списков
    E.append(m * cur_v ** 2 / 2)
    coord_x.append(cur_x)
    velocity_x.append(cur_v)

    # расчет ускорения, скорости и координаты
    F = -(k * cur_x +  b * cur_v)
    acceleration = F / m
    cur_v += (dt * acceleration)
    cur_x += (dt * cur_v)

    # обработка столкновения со стенкой
    if np.abs(cur_x) >= d:
        cnt += 1
        # замена знака скорости - реализация условия упругого соударения
        cur_v *= -1
cnt
```

```python
plt.plot(np.arange(0, 2 + dt, dt), coord_x)
```

```python
plt.plot(np.arange(0, 2 + dt, dt), E)
```
